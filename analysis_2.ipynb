```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
def plot_time_acce(df):
    # Create a line chart
    plt.figure(figsize=(8, 6))
    plt.plot(df['acce_x'], label='x')
    plt.plot(df['acce_y'], label='y')
    plt.plot(df['acce_z'], label='z')

    # Add title and labels
    plt.title('Acceleration')
    plt.xlabel('X-axis Label')
    plt.ylabel('Y-axis Label')

    # Display grid
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
def plot_time_gyro(df):
    # Create a line chart
    plt.figure(figsize=(8, 6))
    plt.plot(df['gyro_x'], label='x')
    plt.plot(df['gyro_y'], label='y')
    plt.plot(df['gyro_z'], label='z')

    # Add title and labels
    plt.title('Gyroscope')

    # Display grid
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
def spectro(x):
# Adjusted sampling frequency
    fs = 100  # Sampling frequency (Hz)
    t = np.arange(0, 5, 1 / fs)  # Time array (10 seconds)

    # Compute the spectrogram
    frequencies, times, Sxx = signal.spectrogram(x, fs, nperseg=64)

    # Plot the spectrogram
    plt.figure(figsize=(10, 6))
    plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
    plt.colorbar(label='Amplitude (dB)')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')
    plt.title('Spectrogram')
    plt.show()
```

```python
df_1 = pd.read_csv('data/right_06.csv')
df_2 = pd.read_csv('data/right_15.csv')
```

### Spectrogram

```python
spectro(df_1['acce_x'])
```

```python
spectro(df_1['gyro_x'])
```
