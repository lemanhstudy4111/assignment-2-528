```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq
```

# Up FFT

```python
df_acce = pd.read_csv('acce_up.csv')
df_gyro = pd.read_csv('gyro_up.csv')
```

```python
x_acce_up = df_acce['x']
y_acce_up = df_acce['y']
z_acce_up = df_acce['z']
```

```python
x_gyro_up = df_gyro['x']
y_gyro_up = df_gyro['y']
z_gyro_up = df_gyro['z']
```

```python
def plot_graph(x_data, y_data, z_data):
    N = 600
    T = 1.0 / 100.0
    xf = fft(x_data)
    yf = fft(y_data)
    zf = fft(z_data)
    dpf = fftfreq(N, T)[:N//2]

    plt.plot(dpf, 2.0/N * np.abs(xf[0:N//2]), label='x')
    plt.plot(dpf, 2.0/N * np.abs(yf[0:N//2]), label='y')
    plt.plot(dpf, 2.0/N * np.abs(zf[0:N//2]), label='z')
    plt.legend()
    plt.xlim(0,20)
    plt.grid()
    plt.show()
```

```python
plot_graph(x_acce_up, y_acce_up, z_acce_up)
```

```python
plot_graph(x_gyro_up, y_gyro_up, z_gyro_up)
```

# Down

```python
df_acce_down = pd.read_csv('acce_down.csv')
df_gyro_down = pd.read_csv('gyro_down.csv')

x_acce_down = df_acce_down['x']
y_acce_down = df_acce_down['y']
z_acce_down = df_acce_down['z']

x_gyro_down = df_gyro_down['x']
y_gyro_down = df_gyro_down['y']
z_gyro_down = df_gyro_down['z']
```

```python
plot_graph(x_acce_down, y_acce_down,z_acce_down)
```

```python
plot_graph(x_gyro_down, y_gyro_down,z_gyro_down)
```
